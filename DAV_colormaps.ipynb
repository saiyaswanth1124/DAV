```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.DataFrame({
    "x":np.random.randn(100),
    "y":np.random.randn(100),
    "values":np.random.randn(100)
})

cmap="coolwarm"
alpha=1
plt.figure(figsize=(6,6))
plt.scatter(data["x"],data["y"],c=data["values"],cmap=cmap,alpha=alpha)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("scatter with colormap")
plt.colorbar(label="value")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


sns.set(rc={"figure.figsize":(6,6)})
```

```python
current_palette=sns.color_palette()
sns.palplot(current_palette)
```

```python
sns.palplot(sns.color_palette('hls',8))
```

```python
sns.palplot(sns.color_palette('husl',8))
```

```python
sample_colors=["windows blue","amber","greyish","faded green","dusty purple",'medium green','denim blue']
sns.palplot(sns.xkcd_palette(sample_colors))
```

```python
sns.palplot(sns.color_palette('cubehelix',8))
```

```python
sns.palplot(sns.cubehelix_palette(8))
```

```python
x,y=np.random.multivariate_normal([0,0],[[1,-.5],[-.5,1]],size=300).T

sample_cmap=sns.cubehelix_palette(light=1,as_cmap=True)
sns.kdeplot(x=x,y=y, cmap=sample_cmap,shade=True)
```

```python
sns.choose_cubehelix_palette(as_cmap=True)
```

```python
sns.palplot(sns.cubehelix_palette(n_colors=8,start=1.7,rot=0.2,dark=.0,light=.95,reverse=True))
```
