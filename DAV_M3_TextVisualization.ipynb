```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from wordcloud import STOPWORDS
```

```python
from google.colab import drive
```

```python
df=pd.read_csv('/content/netflix_titles.csv',usecols=['cast','country'])
df.head()
```

```python
ndf=df.dropna()
ndf.head()
```

```python
text=" ".join(item for item in ndf['cast'])
print(text)
```

```python
stopwords=set(STOPWORDS)
```

```python
wordcloud=WordCloud(background_color="black").generate(text)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.margins(x=0,y=0)
plt.show()
```

None,none,nearest,bilinear,bicubic,spline16,spline36,hanning,kaiser

```python
wordcloud=WordCloud(background_color='white',max_words=100,max_font_size=300,width=800,height=500,colormap='magma').generate(text)
plt.figure(figsize=(20,20))
plt.imshow(wordcloud,interpolation='bicubic')
plt.axis('off')
plt.margins(x=0,y=0)
plt.savefig('cloud.jpg',format='jpg')
plt.show()
```

```python
text1=" ".join(item for item in ndf['country'])
print(text1)
```
